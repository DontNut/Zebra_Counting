```python
import cv2
import imutils
import numpy as np


object_detector = cv2.createBackgroundSubtractorMOG2(detectShadows=True)


cap = cv2.VideoCapture(r'\tmp\Zebra1.mp4')

def Object_Centroid(x,y,w,h):
    x1 = int(w/2)
    y1 = int(h/2)
    x_oc = x+x1
    y_oc = y+y1
    return x_oc , y_oc

while True:
    ret, frame = cap.read()
    if ret == False:
        break
    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey,(3,3),5)
    Zebra_detector = object_detector.apply(blur)
    dilate = cv2.dilate(Zebra_detector, np.ones((5,5)))
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    dilate2 = cv2.morphologyEx(dilate, cv2.MORPH_CLOSE, kernel)
    dilate2 = cv2.morphologyEx(dilate2, cv2.MORPH_CLOSE, kernel)
    contours,h = cv2.findContours(dilate2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.line(frame,(200,0),(200,1200),(225,225,225),3)
        
    counter = 0   
    detect = []
        
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        if cv2.contourArea(contour) < 20000:
            continue
        cv2.rectangle(frame,pt1=(x,y), pt2=(x+w,y+h),color=(0, 255, 0), thickness=2)
        Centroid = Object_Centroid(x,y,w,h)
        detect.append(Centroid)
        print('x:',x,'y:',y,'w:',w,'h:',h)
        #cv2.putText(frame, 'Zebra', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 3)
        cv2.putText(frame, 'Zebra:'+str(counter), (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 3)
        
        for (x,y) in detect:
            if (x<1600) and (y<1600):
                counter += 1
            cv2.line(frame,(200,0),(200,1200),(225,225,225),3)
            detect.remove((x,y))
            print("Zebra:"+str(counter))
    cv2.putText(frame, 'Zebra number:'+str(counter), (1350,100), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3)
        
    cv2.imshow("Zebra_Detector", frame)
    
    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
